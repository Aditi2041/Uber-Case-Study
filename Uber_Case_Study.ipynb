```python
#import numpy , pandas , matplotlib and seaborn packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid") #Setting a seaborn style
```

```python
schedule = pd.read_csv('Uber Request Data.csv') #reading the data
```

```python
schedule
```

Task 1 : Data Cleaning

```python
#Converting Request timestamp and Drop timestamp to datetime format(yyyy-mm-dd hours-minutes-seconds)
schedule["Request timestamp"] = pd.to_datetime(schedule["Request timestamp"] , dayfirst = True)
schedule["Drop timestamp"] = pd.to_datetime(schedule["Drop timestamp"] , dayfirst = True)
```

```python
schedule
```

```python
#Dropping Request id as part of data cleaning as it is redundant to our analysis
schedule = schedule.drop(['Request id'] , axis = 1)
```

```python
schedule
```

```python
#looking at the data types of each column
print(schedule.info())
```

```python
#Differentiating Request timestamp into two columns comprising of date and time
schedule['New_Date'] = [d.date() for d in schedule['Request timestamp']]
schedule['Request_Time'] = [d.time() for d in schedule['Request timestamp']]
```

```python
schedule
```

```python
#Clustering Request time into groups according to different time of day
#Early Morming - 1 A.M to 8 A.M. , Day - 8 A.M. to 12 Noon , Afternoon - 12 Noon to 5 P.M. , Evening - 5 P.M. to 9 P.M. ,
#Late Night - 9 P.M. to 12 P.M.
schedule = schedule.assign(
                 Category_Day = pd.cut(schedule['Request timestamp'].dt.hour,
                 [1 , 7 , 11 , 16 , 21 , 24],
                 labels = [ 'Early Morning' , 'Day' , 'Afternoon' , 'Evening' , 'Late Night']).astype('object'))
```

```python
schedule
```

Subtask 4.1 : Plot Supply Gap for each Pickup point

```python
#Creating dataframe for calculating total demand for each Pickup point
gap = pd.DataFrame(schedule.groupby(['Pickup point'])['Pickup point'].count().rename('Total_Demand')).reset_index()
```

```python
gap
```

```python
#Creating dataframe for calculating total supply for each Pickup point
gap_1 = pd.DataFrame(schedule[schedule['Status'] == 'Trip Completed'].groupby(['Pickup point'])['Pickup point'].count().rename('Total_Supply')).reset_index()
```

```python
gap_1
```

```python
#merging the two dataframes for ease of comparion from a single dataframe
new_gap = pd.merge(gap_1 , gap , how = 'inner')
```

```python
new_gap
```

```python
#plotting total demand and supply for each Pickup point
new_gap.plot(x = 'Pickup point' , kind = 'bar').legend(bbox_to_anchor=(1,1))
```

```python
#creating dataframe for total requests cancelled for each Pickup point
gap_2 = pd.DataFrame(schedule[schedule['Status'] == 'Cancelled'].groupby(['Pickup point'])['Pickup point'].count().rename('Total_Cancelled')).reset_index()
```

```python
gap_2
```

```python
#creating dataframes for requests when no cars were available for each Pickup point 
gap_3 = pd.DataFrame(schedule[schedule['Status'] == 'No Cars Available'].groupby(['Pickup point'])['Pickup point'].count().rename('Total_Unavailable')).reset_index()
```

```python
gap_3
```

```python
#merging the two dataframes gap_2 and gap_3 for calculating the supply gap for each Pickup point
gap_4 = pd.merge(gap_2 , gap_3 , how='inner')
gap_4['Supply_Gap'] = gap_4['Total_Cancelled'] + gap_4['Total_Unavailable']
```

```python
gap_4
```

```python
#merging the dataframes for comparison of supply gap and total supply for each Pickup point
#calculating Difference which how much total cancellations and non-availability of cars is more or less than total trips completed
#for each Pickup point
master_gap = pd.merge(gap_4 , gap_1 , how='inner')
master_gap['Difference'] = master_gap['Supply_Gap'] - master_gap['Total_Supply']
```

```python
master_gap
```

```python
#doing the subplot for supply gap and difference for each Pickup point
plt.figure(figsize=(12 , 7))
plt.subplot(1 ,2 ,1)
plt.title('Supply Gap for Airport and City')
sns.barplot(x='Pickup point' , y='Supply_Gap' , data=master_gap)


plt.subplot(1 , 2 , 2)
plt.title('How much Supply_Gap greater than Total_Supply for Airport and City')
h = sns.barplot(x='Pickup point' , y='Difference' , data=master_gap)
h.set_ylim((0 , 2100))
```

```python
#doing the subplot for total cancellations and non-availability which comprise the Supply gap for each Pickup point for better
#analysis
plt.figure(figsize=(15 , 7))
plt.subplot(1 ,2 ,1)
plt.title('Total Cancellations for Airport and City')
r = sns.barplot(x='Pickup point' , y='Total_Cancelled' , data=master_gap)
r.set_ylim((0 , 1800))


plt.subplot(1 , 2 , 2)
plt.title('Total Unavailable for Airport and City')
sns.barplot(x='Pickup point' , y='Total_Unavailable' , data=master_gap)
```

Subtask 4.2 : Plot Supply Gap for time slots

```python
#creating a dataframe for total number of cancellations for each time of day
category_gap = pd.DataFrame(schedule[schedule['Status'] == 'Cancelled'].groupby(['Pickup point' , 'Category_Day'])['Pickup point'].count().rename('Total_Cancelled')).reset_index()
```

```python
category_gap
```

```python
#creating a dataframe for total number unavailability for each time of day
category_gap_1 = pd.DataFrame(schedule[schedule['Status'] == 'No Cars Available'].groupby(['Pickup point' , 'Category_Day'])['Pickup point'].count().rename('Total_Unavailable')).reset_index()
```

```python
category_gap_1
```

```python
#creating dataframe for total supply which is total trips completed for each time of day
category_gap_2 = pd.DataFrame(schedule[schedule['Status'] == 'Trip Completed'].groupby(['Pickup point' , 'Category_Day'])['Pickup point'].count().rename('Total_Supply')).reset_index()
```

```python
category_gap_2
```

```python
#merging the two dataframes comprising of total cancellations and unavailability to calculate Supply gap for each Time Of Day
category_gap_3 = pd.merge(category_gap , category_gap_1 , how='inner')
category_gap_3['Supply_Gap'] = category_gap_3['Total_Cancelled'] + category_gap_3['Total_Unavailable']
```

```python
category_gap_3
```

```python
#merging the dataframes to calculate total demand and difference for each time of day
#Difference is how much Supply gap(total cancellations and unavailability) is greater or less than total supply(trips completed)
#for each time of day
master_category_gap = pd.merge(category_gap_3 , category_gap_2 , how='inner')
master_category_gap['Total_Demand'] = master_category_gap['Supply_Gap'] + master_category_gap['Total_Supply']
master_category_gap['Difference'] = master_category_gap['Supply_Gap'] - master_category_gap['Total_Supply']
```

```python
master_category_gap
```

```python
#doing subplot for total demand and supply for each time of day
plt.figure(figsize=(20 , 10))
plt.subplot(1 ,2 ,1)
plt.title('Total Demand during different Time Of Day')
sns.barplot(x='Category_Day' , y='Total_Demand' , data=master_category_gap)


plt.subplot(1 , 2 , 2)
plt.title('Total_Supply during different Time Of Day')
g = sns.barplot(x='Category_Day' , y='Total_Supply' , data=master_category_gap)
g.set_ylim((0 , 1950))
```

```python
#doing subplot for Supply gap and difference for each time of day
plt.figure(figsize=(20 , 10))
plt.subplot(1 ,2 ,1)
plt.title('Supply Gap during different Time Of Day')
sns.barplot(x='Category_Day' , y='Supply_Gap' , data=master_category_gap)


plt.subplot(1 , 2 , 2)
plt.title('How much Supply_Gap greater than Total_Supply during different Time Of Day')
s = sns.barplot(x='Category_Day' , y='Difference' , data=master_category_gap)
s.set_ylim((-300 , 1500))
```

```python
#doing subplot for total cancellations and unavailability which comprise the Supply gap for each time of day
plt.figure(figsize=(20 , 10))
plt.subplot(1 ,2 ,1)
plt.title('Total Cancellations during different Time Of Day')
sns.barplot(x='Category_Day' , y='Total_Cancelled' , data=master_category_gap)


plt.subplot(1 , 2 , 2)
plt.title('Total_Unavailable during different Time Of Day')
sns.barplot(x='Category_Day' , y='Total_Unavailable' , data=master_category_gap)
```
